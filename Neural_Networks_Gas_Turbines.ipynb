```python
x= gt2.iloc[:,0:-1]
x
```

```python
y=gt2.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import StandardScaler
a = StandardScaler()
a.fit(x)
X_standardized = a.transform(x)
pd.DataFrame(X_standardized).describe()
```

```python
model = Sequential()
model.add(Dense(12, input_dim=3,  kernel_initializer='uniform', activation='relu'))
model.add(Dense(8,  kernel_initializer='uniform', activation='relu'))
model.add(Dense(1,  kernel_initializer='uniform', activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history= model.fit(x, y,validation_split=0.33, epochs=150, batch_size=10)
```

### Inference:
   We can see that the val_loss i.e., the value of cost function for cross-validation data and loss i.e., the value of cost function for your training data has decreased

```python
_, accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy*100))
```
